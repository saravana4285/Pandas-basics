### <font color="Green"> 4.Finding Missing Values

```python
df.isnull()
```

```python
#Mising Values - column wise
round(100*(df.isnull().sum()/len(df.index)),2)
```

```python
#missing Values row wise
df[df.isnull().sum(axis=1) > 4].head()
```

```python
#createing new datraframe ,excluding the Nan values
new_df = df[df.isnull().sum(axis=1) < 5]
```

```python
new_df.shape
```

```python
new_df.isnull().sum()
```

### <font color="Gold"> 5.Imputing Nan Values 

```python
#imputing with mode value
new_df.loc[pd.isnull(new_df['Purchase Address']),['Purchase Address']] = new_df['Purchase Address'].mode()[0]
```

```python
new_df.isnull().sum()
```

```python
#imputing with Mean Value
new_df.loc[pd.isnull(new_df['Price Each']),['Price Each']] = new_df['Price Each'].mode()[0]
```

```python
new_df.isnull().sum()
```
